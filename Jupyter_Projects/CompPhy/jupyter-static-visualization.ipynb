## Python Visualization Libraies

```python
import pretty_errors
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Matplotlib

```python
np.random.seed(0)

mu = 200
sigma = 25
x = np.random.normal(mu, sigma, size = 100)

fig, (ax0, axl) = plt.subplots(ncols = 2, figsize = (8, 4))

ax0.hist(x, 20, density = 1, histtype = 'stepfilled', facecolor = 'g', alpha = 0.75)
ax0.set_title('stepfilled')

# Create a histogram by providing the bin edges (unequally spaced).
bins = [100, 150, 180, 195, 205, 220, 250, 300]
axl.hist(x, bins, density = 1, histtype = 'bar', rwidth = 0.8)
axl.set_title('unequal bins')

fig.tight_layout()
plt.show()
```
